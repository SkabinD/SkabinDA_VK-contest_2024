# Imports

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages
import seaborn as sns
```

```python
PATH_TRAIN = "../data/train_df.csv"
PATH_TEST  = "../data/test_df.csv"
PATH_IMAGES = "../images/"
```

```python
df_train = pd.read_csv(PATH_TRAIN)
df_test  = pd.read_csv(PATH_TEST)
```

## Матрица корреляций

```python
%%capture

pdf = PdfPages(PATH_IMAGES+"corr_mtx.pdf", )

corr_mtx = df_train.corr()

fig = plt.figure(figsize=(40, 40))

sns.heatmap(corr_mtx, annot=True)

plt.tight_layout()
plt.show()

pdf.savefig(fig)
pdf.close()
```

## Распределения

```python
%%capture
pdf = PdfPages(PATH_IMAGES+"histplots.pdf", )

cols = df_train.columns[1:]

for c in cols:
    fig = plt.figure()
    sns.histplot(df_train, x=c, label="train")
    sns.histplot(df_test, x=c, label="test")
    plt.legend()
    plt.show()
    pdf.savefig(fig)

pdf.close()
```

```python
%%capture
pdf = PdfPages(PATH_IMAGES+"histplots_prob.pdf", )

cols = df_train.columns[1:]

for c in cols:
    fig = plt.figure()
    sns.histplot(df_train, x=c, label="train", stat="probability", kde=True)
    sns.histplot(df_test, x=c, label="test", stat="probability", kde=True)
    plt.legend()
    plt.show()
    pdf.savefig(fig)

pdf.close()
```

```python
sns.scatterplot(df_train, x="feature_1", y="target", label="train")
print(df_train["feature_1"].unique())
sns.scatterplot(df_test, x="feature_1", y="target", label="test")
print(df_test["feature_1"].unique())
plt.show()
```
